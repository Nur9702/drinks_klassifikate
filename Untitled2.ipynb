```python
from fastai.vision.all import *
from ipywidgets import widgets


#PATH
path = Path('train')
#print('path')

fls = get_image_files(path)
print(fls) # Check if it is correctly finding image files with supported file extensions
print(len(fls)) # Check if the length of fls is appropriate
failed = verify_images(path)
print(failed) #Check and fix or delete corrupted images

#Datablock yaratamiz
transports = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(224))

#data loader

dls = transports.dataloaders(path)

#dls.train.show_batch(max_n=32, nrows=4)
#O'qitish
learn = cnn_learner(dls, resnet34, metrics=accuracy)
learn.fine_tune(4)
```

```python
#aniqlikni tekshiramiz
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(5, nrows=1)
```
